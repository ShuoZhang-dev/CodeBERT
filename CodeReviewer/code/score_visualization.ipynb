```python
import json
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
result_by_conf_score_path = os.path.join("/datadisk/shuo/CodeReview/CodeReviewer/test_result/cls", "test_result_by_threshold.json")
with open(result_by_conf_score_path, 'r') as f:
    all_results = json.load(f)

thresholds, precisions, recalls, f1_scores, accuracies = [], [], [], [], []
for thred, report_dict in all_results.items():
    thresholds.append(float(thred))
    precisions.append(report_dict["1"]["precision"])
    recalls.append(report_dict["1"]["recall"])
    f1_scores.append(report_dict["1"]["f1-score"])
    accuracies.append(report_dict["accuracy"])

data_labels = ["threshold="+str(t) for t in thresholds]
fig, ax = plt.subplots(figsize=(12,8))
plt.plot(recalls, precisions, 'bo--')
plt.xlabel("recall")
plt.ylabel("precision")
plt.title("Precision-Recall Curve for T5 Quality Estimation with Different Confidence Threshold")
for i in range(len(recalls)):
    ax.text(recalls[i], precisions[i], data_labels[i], size=8)
plt.show()
```
